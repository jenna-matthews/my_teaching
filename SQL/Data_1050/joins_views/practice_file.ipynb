```python
# course table
query = """ 
        SELECT *
        from courses
        limit 10
            ;
        """

result_course = sqldf(query)
result_course
```

```python
# course table
query = """ 
        SELECT *
        from assessments a 
        full outer join courses c on a.code_presentation = c.code_presentation
        limit 10
            ;
        """

result_assess = sqldf(query)
result_assess
```

```python
# course table
query = """ 
        CREATE or REPLACE 
            VIEW vw_assessments AS
        SELECT *
        from assessments
        LEFT OUTER JOIN courses
        ON assessments.code_presentation 
            = courses.code_presentation
        limit 10
            ;
        """

result_assess = sqldf(query)
result_assess
```
